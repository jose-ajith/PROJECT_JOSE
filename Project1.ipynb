```python
from datetime import datetime

books = {
    "B1": {"title": "Python Basics", "author": "John", "total": 3, "qty": 3},
    "B2": {"title": "Data Structures", "author": "Jane", "total": 2, "qty": 2}
}
students = {
    "S1": {"name": "Jose", "borrowed": {}},
    "S2": {"name": "Christo", "borrowed": {}},
    "S3": {"name": "Joseph", "borrowed": {}}
}

def issue_book(stud_id, bk_id):
    if stud_id not in students or bk_id not in books:
        return print("Error: Invalid Student or Book ID.")

    student, book = students[stud_id], books[bk_id]

    if book["qty"] <= 0:
        print("Error: No copies available.")
    elif len(student["borrowed"]) >= 3:
        print("Error: Max limit (3) reached.")
    elif bk_id in student["borrowed"]:
        print("Error: Already issued to this student.")
    else:
        book["qty"] -= 1
        student["borrowed"][bk_id] = datetime.now()
        print(f"Success: {book['title']} issued to {student['name']}.")

def return_book(stud_id, bk_id, days_kept):
    if stud_id not in students or bk_id not in students[stud_id]["borrowed"]:
        return print("Error: Borrow record not found for this student and book.")

    books[bk_id]["qty"] += 1
    del students[stud_id]["borrowed"][bk_id]

    penalty = max(0, (days_kept - 14) * 10)
    print(f"Success: Book returned. Days kept: {days_kept}. Penalty: {penalty}rs")

def search_books(query):
    print(f"\n--- Results for '{query}' ---")
    for b_id, info in books.items():
        if query.lower() in info["title"].lower() or query.lower() in info["author"].lower():
            print(f"[{b_id}] {info['title']} by {info['author']} ({info['qty']}/{info['total']} available)")

def display_all_books():
    print("\n=== COMPLETE BOOK INVENTORY ===")
    print(f"{'Book ID':<10}{'Title':<25}{'Author':<15}{'Available/Total'}")
    print("-" * 65)
    for b_id, info in books.items():
        print(f"{b_id:<10}{info['title']:<25}{info['author']:<15}{info['qty']}/{info['total']}")

def display_all_students():
    print("\n=== REGISTERED STUDENTS ===")
    print(f"{'Student ID':<15}{'Student Name':<25}{'Books Borrowed'}")
    print("-" * 55)
    for s_id, info in students.items():
        borrowed_count = len(info["borrowed"])
        print(f"{s_id:<15}{info['name']:<25}{borrowed_count} / 3")

while True:
    print("\n1. Issue Book      2. Return Book      3. Search Books")
    print("4. View All Books  5. View All Students 6. Exit")
    choice = input("Select option: ").strip()

    if choice == "1":
        issue_book(input("Student ID: "), input("Book ID: "))
    elif choice == "2":
        s_id = input("Student ID: ")
        b_id = input("Book ID: ")
        try:
            days = int(input("Enter number of days kept: "))
            return_book(s_id, b_id, days)
        except ValueError:
            print("Error: Days kept must be a valid number.")
    elif choice == "3":
        search_books(input("Search keyword: "))
    elif choice == "4":
        display_all_books()
    elif choice == "5":
        display_all_students()
    elif choice == "6":
        print("Exiting system.")
        break
    else:
        print("Invalid choice.")
```
